```python
from astropy.table import vstack, hstack, Table
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.io import ascii, fits
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
M67 = Table.read('Downloads/M67.fits')
```

```python
M67.info()
```

```python
g = M67['psfMag_g']
i = M67['psfMag_i']
r = M67['psfMag_r']
u = M67['psfMag_u']
z = M67['psfMag_z']
RA = M67['ra']
DEC = M67['dec']
u_g = u-g
r_i = r-i
r_z = r-z
g_i = g-i
```

```python
lilM67 = Table.read('Downloads/lilM67.fits')
g2 = lilM67['psfMag_g']
i2 = lilM67['psfMag_i']
r2 = lilM67['psfMag_r']
u2 = lilM67['psfMag_u']
z2 = lilM67['psfMag_z']
RA2 = lilM67['ra']
DEC2 = lilM67['dec']
u_g2 = u2-g2
r_i2 = r2-i2
r_z2 = r2-z2
g_i2 = g2-i2
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
fig.tight_layout()

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.gca().invert_xaxis()
ax.scatter(u_g, g , s=46, color='cyan', alpha = 0.5, marker = '.', label='All Stars')
ax.scatter(u_g2, g2 , s=46, color='orange', alpha = 0.5, marker = '.', label='Inner Stars')
ax.set_ylabel('g', fontsize=28)
ax.set_xlabel('u-g', fontsize=28) 
plt.legend(loc='upper left', fontsize=20, markerscale=6)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
fig.tight_layout()

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.gca().invert_xaxis()
ax.scatter(r_i, r , s=46, color='cyan', alpha = 0.5, marker = '.', label='All Stars')
ax.scatter(r_i2, r2 , s=46, color='orange', alpha = 0.5, marker = '.', label='Inner Stars')
ax.set_ylabel('r-i', fontsize=28)
ax.set_xlabel('r', fontsize=28) 
plt.legend(loc='upper left', fontsize=20, markerscale=3)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
fig.tight_layout()

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
ax.scatter(u_g, r_i , s=46, color='cyan', alpha = 0.5, marker = '.', label='All Stars')
ax.scatter(u_g2, r_i2 , s=46, color='orange', alpha = 0.5, marker = '.', label='Inner Stars')
ax.set_ylabel('r-i', fontsize=28)
ax.set_xlabel('u-g', fontsize=28) 
plt.legend(loc='upper left', fontsize=20, markerscale=6)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
fig.tight_layout()

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
ax.scatter(g_i, r_z , s=46, color='cyan', alpha = 0.5, marker = '.', label='All Stars')
ax.scatter(g_i2, r_z2 , s=46, color='orange', alpha = 0.5, marker = '.', label='Inner Stars')
ax.set_ylabel('r-z', fontsize=28)
ax.set_xlabel('g-i', fontsize=28) 
plt.legend(loc='upper left', fontsize=20, markerscale=6)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
fig.tight_layout()

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
ax.scatter(RA, DEC , s=46, color='cyan', marker = '.', label='All Stars')
ax.scatter(RA2, DEC2 , s=46, color='orange', marker = '.', label='Inner Stars')
ax.set_ylabel('RA', fontsize=28)
ax.set_xlabel('Dec', fontsize=28)
plt.legend(loc='upper left', fontsize=20, markerscale=3)
```
