```python
#create sine function for curve fit
#a is amplitude, b is frequency
def SineX(x,a,b,c,d):
    return (a*np.sin((b*x)+c))+d
```

```python
#C-Rich Mira
StarCRM = pd.read_csv('Desktop/New Data/Light Curves/OGLE-LMC-LPV-50426.csv')
DateCRM = StarCRM['HJD-2450000'][:]
MagCRM = StarCRM['IMag'][:]

DateMin = min(DateCRM)
DateMax = max(DateCRM)

GuessAA = 0.75
GuessBB = 0.0165
GuessCC = 1.75
GuessDD = 15.3
Guesses2 = [GuessAA, GuessBB, GuessCC, GuessDD]

fitpars2, error = curve_fit(SineX,DateCRM,MagCRM,p0=Guesses2)

ZCRM = np.linspace(0, 5000, len(DateCRM))
```
